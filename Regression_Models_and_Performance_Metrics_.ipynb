## Supervised Learning: Regression Models and Performance Metrics

1. What is Simple Linear Regression (SLR)? Explain its purpose.
  - Simple Linear Regression is a statistical method used to understand the relationship between one independent variable and one dependent variable. It helps us see how a change in one variable affects another. The main purpose of SLR is to predict future values and identify trends using a straight-line relationship.

2. What are the key assumptions of Simple Linear Regression?
  - The key assumptions of Simple Linear Regression are:
     - Linearity: There should be a straight-line relationship between the independent variable and the dependent variable.
     - Independence: Each observation should be independent, meaning the value of one data point should not affect another.
     - Homoscedasticity: The variance of errors should remain constant for all values of the independent variable.
     - Normality of errors: The residuals (errors) should follow a normal distribution.
     - No extreme outliers: There should not be extreme outliers, as they can strongly affect the regression line.

3. Write the mathematical equation for a simple linear regression model and
explain each term.
  - The equation is:   y = mx + c
  - Where:
     - y is the dependent variable
     - x is the independent variable
     - m is the slope (shows the change in y for one unit change in x)
     - c is the intercept (value of y when x is zero)

4.  Provide a real-world example where simple linear regression can be
applied.
  - Simple linear regression can be used to predict house prices based on size. Here, house size is the independent variable and price is the dependent variable. As size increases, the price usually increases.
  - One more example- A student's exam score based on the number of hours studied. Here, hours studied is the independent variable, and exam score is the dependent variable. As study hours increase, the exam score generally increases, showing a linear relationship.

5.  What is the method of least squares in linear regression?
  - The method of least squares is used to find the best-fit line by minimizing the sum of squared differences between actual values and predicted values. It ensures the line fits the data as closely as possible.

6. What is Logistic Regression? How does it differ from Linear Regression?
  - Logistic Regression is a supervised learning algorithm used for classification problems.
  - It predicts probability values between 0 and 1.
  - It uses a sigmoid function to convert outputs into probabilities.
  - The output is categorical, such as Yes/No or Pass/Fail.
  - Difference between them:
       - Linear Regression is used to predict continuous values like price or temperature.
       - Logistic Regression is used to predict categories or classes.
       - Linear Regression outputs any real value, while Logistic Regression outputs values between 0 and 1.

7. Name and briefly describe three common evaluation metrics for regression
models.
  - Mean Absolute Error (MAE): Average of absolute differences between actual and predicted values.
  - Mean Squared Error (MSE): Average of squared differences between actual and predicted values.
  - Root Mean Squared Error (RMSE): Square root of MSE, shows error in original units.

8. What is the purpose of the R-squared metric in regression analysis?
  - R-squared is used to measure how well a regression model explains the variation in the dependent variable. It shows the proportion of the total variation that is explained by the independent variable. Its value lies between 0 and 1, where a higher value indicates a better fit of the model.

9. How do you interpret the coefficients in a simple linear regression model?
  - In a simple linear regression model, the coefficient of the independent variable (slope) shows how much the dependent variable changes when the independent variable increases by one unit. A positive slope means the variables increase together, while a negative slope means one decreases as the other increases. The intercept represents the value of the dependent variable when the independent variable is zero.


```python
# 10. Write Python code to fit a simple linear regression model using scikit-learn and print the slope and intercept.
# (Include your Python code and output in the code box below.)

import numpy as np
from sklearn.linear_model import LinearRegression

# Here, X is the independent variable and y is the dependent variable.
# reshape(-1, 1) is used because scikit-learn expects input data in 2D format.

X = np.array([1, 2, 3, 4, 5]).reshape(-1, 1)
y = np.array([2, 4, 6, 8, 10])

model = LinearRegression()
model.fit(X, y)

print("Slope:", model.coef_[0])
print("Intercept:", model.intercept_)

# Explanation of Output:
# The slope (2.0) means that for every 1-unit increase in X, the value of y increases by 2 units.
# The intercept (0.0) means that when X is 0, the predicted value of y is 0.
# This shows a perfect linear relationship between X and y.
```
